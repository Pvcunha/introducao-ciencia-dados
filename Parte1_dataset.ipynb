```python
#remove as duas primeiras colunas -> Unnamed e id
data_train = data_train.iloc[: , 1:]
data_test = data_test.iloc[: , 1:]
data_train.head()
```

```python
vars_numerical = ['id', 'Age', 'Flight_Distance', 'Departure_Delay_in_Minutes', 'Arrival_Delay_in_Minutes']
vars_categorical = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Inflight wifi service', 'Departure/Arrival time convenient',
'Ease of Online booking', 'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment', 'On-board service',
 'Leg room service', 'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness', 'satisfaction']
```

```python
data = data.rename(columns = {"Flight Distance": "Flight_Distance", "Departure Delay in Minutes": "Departure_Delay_in_Minutes", 'Arrival Delay in Minutes': 'Arrival_Delay_in_Minutes'})
data[vars_numerical].describe()
```

```python
data_ = data[vars_numerical].describe().T.rename_axis('column_name').reset_index()
data_
```

Calculating IQR

```python
data_['iqr'] = data_['75%'] - data_['25%']
```

Calculating Skewness and Kurtosis

```python
data_skewness =  pd.DataFrame(data[vars_numerical].skew(skipna = True).rename_axis('column_name').reset_index().rename(columns = {0:'skewness'}))
data_kurtosis = pd.DataFrame(data[vars_numerical].kurtosis(skipna = True).rename_axis('column_name').reset_index().rename(columns = {0:'skewness'}))
data_kurtosis
```

Adicionando esses calculos ao sumario 

```python
data_ = ps_train_summary = data_.merge(data_skewness, on = 'column_name').merge(data_kurtosis, on = 'column_name')
data_
```

Usando as regras de tukey's outlier

```python
data_['lower_bound'] = data_['25%'] - (1.5 * data_['iqr'])
data_['upper_bound'] = data_['75%'] + (1.5 * data_['iqr'])
data_
```

```python
for column in vars_numerical:
    plt.figure(figsize = (10,2))
    sns.boxplot(data = data, x = column)
```

Visualmente, há um atraso em Departure_Delay_in_Minutes and Arrival_Delay_in_Minutes

```python
flight_distance_upper = data_[data_['column_name'] == 'Flight_Distance'][['upper_bound']].iloc[0,0]
flight_distance_lower = data_[data_skewness['column_name'] == 'Flight_Distance'][['lower_bound']].iloc[0,0]

print("Flight Distance upper: {}".format(flight_distance_upper))
print("Flight Distance lower: {}".format(flight_distance_lower))

flight_distance_outlier = data[(data["Flight_Distance"] < flight_distance_lower) | 
                       (data["Flight_Distance"] > flight_distance_upper)]

print(flight_distance_outlier['Flight_Distance'])
flight_distance_outlier.groupby('Flight_Distance')['Flight_Distance'].count()
```
